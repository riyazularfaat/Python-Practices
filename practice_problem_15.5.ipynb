## 1. Add New Column and Save File‎
You have a DataFrame with column Price. 

Create a new column Discounted_Price with (10% discounted price). 

Save the modified DataFrame as discounted_products.csv.‎
________________________________________


```python
import pandas as pd
```

```python
df = pd.read_csv("practice_data.csv", on_bad_lines = 'skip') # on_bad_lines is used to skip any problematic lines (like line 3 in your case) that don’t match the expected number of fields.
df['Discounted_Price'] = df['Price'] - df['Price'] * 0.1
df.to_csv('discounted_products.csv')
```

## 2. Check Duplicates and Remove Them‎
Column: EmployeeID

* Count duplicate entries

```python
df = pd.read_csv('discounted_products.csv')
duplicated_count = df['EmployeeID'].duplicated().sum()
print(duplicated_count)
```

## 3. Handle Missing Values
Columns: Age, Salary 
 
 Tasks:
* Count missing values 

* Fill missing Age with mean age 

* Drop rows where Salary is missing


```python
count_null_age = df['Age'].isnull().sum()
count_null_salary = df['Salary'].isnull().sum()
print(f"Null Count of Age: {count_null_age}")
print(f"Null Count of Salary {count_null_salary} ")
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.dropna(subset = ['Salary'], inplace=True)
```

```python
df
```

## 4. Apply Statistical Functions
Columns: Marks1, Marks2, Marks3 

Tasks:
* Calculate mean, max for each subject 

* Compute total marks per student 

* Find standard deviation of total marks


```python
marks1_mean =df['Marks1'].mean()
marks1_max =df['Marks1'].max()
print(f'Marks1: mean => {marks1_mean}, max => {marks1_max} ')
marks2_mean =df['Marks2'].mean()
marks2_max =df['Marks2'].max()
print(f'Marks2: mean => {marks2_mean}, max => {marks2_max} ')
marks3_mean =df['Marks3'].mean()
marks3_max =df['Marks3'].max()
print(f'Marks3: mean => {marks3_mean}, max => {marks3_max} ')
```

```python
df['Total Marks'] = df[['Marks1', 'Marks2', 'Marks3']].sum(axis = 1)
df
```

```python
df['Total Marks'].std()
```

## 5. Histogram, Bar, Scatter, Pie‎
Create:‎
* Histogram of Age
* Scatter plot (Age vs Salary)‎
* Scatter plot ( Price vs Discounted Price)  


```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df["Age"], color='teal', edgecolor = 'black')

plt.title("Age of students")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df['Age'], df['Salary'], color = 'green')

plt.title("Correlation between Age and Salary of students")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(df['Price'], df['Discounted_Price'], color='red')

plt.title("Correlation between Price and Discounted Price")
plt.xlabel("Price")
plt.ylabel("Discounted_Price")
plt.show()
```
